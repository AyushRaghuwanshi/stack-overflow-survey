## import dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('survey_results_public.csv')
```

```python
print(df.shape)
df.head()
```

## data exploration

```python
# col names which dont have any nan values
df.isnull().sum()[df.isnull().sum() == 0].index
```

```python
# persons who thinks OSS is, on average, of HIGHER quality than proprietary / closed source
df_oss_is_better = df[df['OpenSource'] == 'OSS is, on average, of HIGHER quality than proprietary / closed source software']
print(df_oss_is_better.shape)
print((df_oss_is_better.shape[0]/df.shape[0])*100)
df_oss_is_better.head()
```

```python
# persons who thinks OSS is, on average, of LOWER quality than proprietary / closed source software
df_oss_is_lower = df[df['OpenSource'] == 'OSS is, on average, of LOWER quality than proprietary / closed source software']
print(df_oss_is_lower.shape)
print((df_oss_is_lower.shape[0]/df.shape[0])*100)
df_oss_is_lower.head()
```

```python
##persons those are students
df_student_true = df[df['Student'].isin(['Yes, full-time','Yes, part-time'])]
print(df_student_true.shape)
df_student_true.head()
```

```python
#students who thinks oss quality is not that good
df_student_true_oss_lower = df_student_true[df_student_true['OpenSource'] == 'OSS is, on average, of LOWER quality than proprietary / closed source software']
print(df_student_true_oss_lower.shape)
print((df_student_true_oss_lower.shape[0]/df_student_true.shape[0])*100)
df_student_true_oss_lower.head()
```

```python
#
code_first = df[df['Age1stCode'] < '15']

code_first.Age1stCode.head()
```

```python
code_first.head()
```

```python
code_first_count = code_first.groupby('Country')
```

```python
code_first_count.count().head()
```

```python
code_first_count.count().sort_values(by = 'Respondent', ascending = False).head(10)['Respondent']
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([2,2,3,3])
country = code_first_count.count().sort_values(by = 'Respondent', ascending = False).head(10)['Respondent'].index
number_of_coders = code_first_count.count().sort_values(by = 'Respondent', ascending = False).head(10)['Respondent'].values
ax.bar(country,number_of_coders)
plt.show()
```
